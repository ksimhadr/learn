Data Visualization

t-SNE
t-distributed Stochastic Neighbor Embedding

- visualizing high dimensional data
- unsupervised non-linear dimensionality reduction technique for data exploration and visualization

```python
from IPython.display import Image
#Image(filename="tSNE.png", embed=True)
#Image(filename="tSNE-gif.gif", embed=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets #for dataset
from sklearn.manifold import TSNE #import TSNE from sklearn.manifold

#Load sample dataset
iris = datasets.load_iris()
x = iris.data
y = iris.target

print(x.shape)
print(y.shape)

print(x[:5,:]) #considering whole data, 4-D data is being used

tsne = TSNE(n_components=2, random_state=400) #2D, randomly distributed data
x_tsne = tsne.fit_transform(x) #4D to 2D

print(x_tsne.shape)

#Plot the t-SNE results
plt.figure(figsize = (8, 6))
plt.scatter(x_tsne[:, 0], x_tsne[:, 1], c=y, cmap=plt.cm.get_cmap("jet", 10), edgecolor ='k') #colour y samples
#plt.colorbar(ticks=range(10))
#plt.clim(-0.5, 9.5)
plt.title("t-SNE visualization of the iris dataset")
plt.xlabel("t-SNE Feature 1")
plt.ylabel("t-SNE Feature 2")
plt.show()

tsne2 = TSNE(n_components=3, random_state=400) #3D, randomly distributed data
x_tsne2 = tsne2.fit_transform(x) #4D to 3D

print(x_tsne2.shape)
#Plot the t-SNE2 results in 3D
plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(x_tsne2[:, 0], x_tsne2[:, 1], x_tsne2[:, 2], c=y, cmap=plt.cm.get_cmap("jet", 10), edgecolor ='k') #colour y samples
plt.title("t-SNE2 visualization of the iris dataset")
plt.xlabel("t-SNE2 Feature 1")
plt.ylabel("t-SNE2 Feature 2")
plt.show()

```

Digits data visualization

- Objective: Visualize the high-dimensional MNIST digit dataset (28x28 pixel images) in a lower-dimensional space for intuitive understanding.
-
Dataset: MNIST dataset contains 70,000 grayscale images of handwritten digits (0 through 9), with 60,000 for training and 10,000 for testing. -

Data Preprocessing: Flatten each image into a 1D array, resulting in a 784-dimensional feature vector for each digi - t.

t-SNE Application: Utilize the t-distributed Stochastic Neighbor Embedding (t-SNE) algorithm to reduce the dimensionality of the feature vectors to - E space.

Color Mapping: Assign different colors to each digit class, facilitating class differentiation in the visudimensional space.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.manifold import TSNE

# Load the MNIST digits dataset
digits = datasets.load_digits()
X = digits.data
y = digits.target

# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)#MNIST  784 D data to 2 component or 2D data

# Plot the t-SNE results
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap=plt.cm.tab10, edgecolor='k')
plt.title("t-SNE Visualization of MNIST Digits")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.manifold import TSNE

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Apply t-SNE with 3 components for 3D visualization
tsne = TSNE(n_components=3, random_state=42)
X_tsne = tsne.fit_transform(X)

# Plot the t-SNE results in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(X_tsne[:, 0], X_tsne[:, 1], X_tsne[:, 2], c=y, cmap=plt.cm.Set1, edgecolor='k')

ax.set_title("t-SNE 3D Visualization of Iris Dataset")
ax.set_xlabel("t-SNE Feature 1")
ax.set_ylabel("t-SNE Feature 2")
ax.set_zlabel("t-SNE Feature 3")

# Add a colorbar
colorbar = fig.colorbar(scatter, ax=ax, pad=0.1, fraction=0.05)
colorbar.set_label("Class")

plt.show()
```

##IRIS dataset revisit

1. The Iris dataset consists of three species: setosa, versicolor, and virginica.
2. The t-SNE algorithm is used to reduce the dimensionality of the dataset to 2D
3. The resulting plot shows how the different species cluster in the reduced-dimensional space.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.manifold import TSNE

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# Apply t-SNE with 2 components for 2D visualization
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Plot the t-SNE results
plt.figure(figsize=(10, 8))
colors = ['red', 'green', 'blue']

for i, c in zip(range(len(target_names)), colors):
    plt.scatter(X_tsne[y == i, 0], X_tsne[y == i, 1], c=c, label=target_names[i])

plt.title("t-SNE Visualization of Iris Dataset")
plt.xlabel("t-SNE Feature 1")
plt.ylabel("t-SNE Feature 2")
plt.legend()
plt.show()
```
