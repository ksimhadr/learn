```python
#matplotlib and seaborn - data visualization
#pandas - data extraction (excel/tabular form data)
#numpy - numerical python
#plotly - gui based plots
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import plotly.express as px
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data_iris = sns.load_dataset("iris")
print(data_iris)
sns.lineplot(x = 'sepal_length', y = 'sepal_width',data=data_iris,color = 'green')
plt.title("Sepal length vs Sepal Width of IRIS")
plt.plot()
```

```python
sns.boxplot(x = 'sepal_length', y = 'sepal_width',data=data_iris)
plt.title("Box Plot - Sepal length vs Sepal Width of IRIS")
plt.plot()
```
